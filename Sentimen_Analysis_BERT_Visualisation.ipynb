```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('youtube_reviews_english_2025.csv')
df.head()
```

```python
df.shape[0]
```

```python
month_counts = df['month'].value_counts().sort_index()

month_names = [
    "Jan", "Feb", "Mar", "Apr", "May", "Jun",
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
]

plt.figure(figsize=(10,5))
bars = plt.bar(month_names, month_counts.values)

plt.title("Number of Reviews Per Month")
plt.xlabel("Month")
plt.ylabel("Number of Reviews")
plt.bar_label(bars, padding=3)

plt.show()
```

```python
target_words = ["app", "youtube", "channel", "video", "tube", "utube"]

text_data = df["review_description"].fillna("").str.lower()
word_counts = pd.Series({
    word: text_data.str.count(rf"\b{word}\b").sum()
    for word in target_words
}).sort_values(ascending=False)

plt.figure(figsize=(9, 5))
bars = plt.bar(word_counts.index, word_counts.values)

plt.title("Number of Word Occurrences in YouTube Reviews")
plt.xlabel("Word")
plt.ylabel("Number of Occurrences")
plt.bar_label(bars, padding=3)
plt.tight_layout()
plt.show()

word_counts
```
