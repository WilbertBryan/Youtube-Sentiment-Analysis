```python
import pandas as pd

df = pd.read_csv('/content/youtube_reviews_2025_all.csv')
df.head()
```

```python
df.info()
```

**Collect English Dataset**

```python
from langdetect import detect, DetectorFactory
DetectorFactory.seed = 0

def is_english(text):
    try:
        return detect(text) == 'en'
    except:
        return False

df = df.dropna(subset=['review_description'])
df_en = df[df['review_description'].apply(is_english)]

print("Jumlah data Inggris:", len(df_en))
```

```python
df_en
```

```python
df_en.to_csv('youtube_reviews_english_2025.csv', index=False)
print("CSV berhasil disimpan.")
```

**Filter Reviews with Minimum length 20 & Collect 500 data each Month**

```python
# Filter review terlalu pendek (misal <20 karakter)
MIN_REVIEW_LENGTH = 20
df_minLength = df_en[df_en['review_description'].str.len() >= MIN_REVIEW_LENGTH]

# Sampling 500 review per bulan
SAMPLE_PER_MONTH = 500

df_sample = (
    df_minLength.groupby('month', group_keys=False)
      .apply(lambda x: x.sample(
          n=min(SAMPLE_PER_MONTH, len(x)),
          random_state=42
      ))
)

df_sample.reset_index(drop=True, inplace=True)

# Cek distribusi per bulan
print(df_sample['month'].value_counts().sort_index())
```

```python
df_sample
```

```python
df_sample.to_csv('youtube_reviews_sampleenglish_2025.csv', index=False)
print("youtube_reviews_english_2025.csv berhasil disimpan.")
```
